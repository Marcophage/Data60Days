```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend. figure_format = 'retina'
# 準備數據 ... 假設我要畫一個cos波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,3 * np.pi ,0.1)
y = np.cos(x)
# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 cos 圖形

plt.savefig("cos.png",dpi=300,format="png")
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)


plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.plot('r')
plt.scatter(X, Y, marker='o', c='c',alpha=0.5,label='sample')
plt.title("Scatter plot")
plt.legend( loc= 'upper right')

plt.show()
```
